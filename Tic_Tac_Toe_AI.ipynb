#TIC-TAC-TOE


```python
import math

def print_board(board):
    print("\nCurrent Board:")
    for row in board:
        print(" | ".join(row))
        print("-" * 5)

def check_winner(board):
    for row in board:
        if row.count(row[0]) == 3 and row[0] != ' ':
            return row[0]
    for col in range(3):
        if (board[0][col] == board[1][col] == board[2][col]) and board[0][col] != ' ':
            return board[0][col]
    if (board[0][0] == board[1][1] == board[2][2]) and board[0][0] != ' ':
        return board[0][0]
    if (board[0][2] == board[1][1] == board[2][0]) and board[0][2] != ' ':
        return board[0][2]
    return None

def is_full(board):
    return all(cell != ' ' for row in board for cell in row)

def minimax(board, is_max):
    winner = check_winner(board)
    if winner == 'O': return 1
    if winner == 'X': return -1
    if is_full(board): return 0

    if is_max:
        best = -math.inf
        for i in range(3):
            for j in range(3):
                if board[i][j] == ' ':
                    board[i][j] = 'O'
                    score = minimax(board, False)
                    board[i][j] = ' '
                    best = max(score, best)
        return best
    else:
        best = math.inf
        for i in range(3):
            for j in range(3):
                if board[i][j] == ' ':
                    board[i][j] = 'X'
                    score = minimax(board, True)
                    board[i][j] = ' '
                    best = min(score, best)
        return best

def best_move(board):
    best_val = -math.inf
    move = (-1, -1)
    for i in range(3):
        for j in range(3):
            if board[i][j] == ' ':
                board[i][j] = 'O'
                move_val = minimax(board, False)
                board[i][j] = ' '
                if move_val > best_val:
                    move = (i, j)
                    best_val = move_val
    return move

def get_valid_input(board):
    while True:
        try:
            row = int(input("Enter row (0-2): "))
            col = int(input("Enter col (0-2): "))
            if row not in range(3) or col not in range(3):
                print("âš ï¸ Invalid position. Please enter values between 0 and 2.")
            elif board[row][col] != ' ':
                print("âš ï¸ That spot is already taken. Try again.")
            else:
                return row, col
        except ValueError:
            print("âš ï¸ Please enter valid numbers (0, 1, or 2).")

def play_game():
    board = [[' ' for _ in range(3)] for _ in range(3)]
    turn = 1

    while True:
        print(f"\nğŸ® Turn #{turn}")
        print_board(board)

        print("ğŸ§‘ Your move!")
        row, col = get_valid_input(board)
        board[row][col] = 'X'

        if check_winner(board) or is_full(board):
            break

        print("ğŸ¤– AI is thinking...")
        ai_row, ai_col = best_move(board)
        board[ai_row][ai_col] = 'O'

        if check_winner(board) or is_full(board):
            break

        turn += 1

    print_board(board)
    result = check_winner(board)
    if result == 'X':
        print("ğŸ‰ You win!")
    elif result == 'O':
        print("ğŸ’» AI wins! Better luck next time.")
    else:
        print("ğŸ¤ It's a draw!")

# Main Loop
while True:
    play_game()
    replay = input("\nğŸ” Do you want to play again? (Y/N): ").strip().lower()
    if replay != 'y':
        print("ğŸ‘‹ Thanks for playing!")
        break
```
